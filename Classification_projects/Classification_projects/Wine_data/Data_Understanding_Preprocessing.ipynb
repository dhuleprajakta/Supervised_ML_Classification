```python
j=1
plt.figure(figsize=(25,20))
for i in continous:
    plt.subplot(4,4,j)
    sns.boxplot(df[i])
    plt.title(f'{i}')
    j=j+1
plt.show()
```

**Too few outliers that can be negligible**

```python
j=1
plt.figure(figsize=(25,20))
for i in continous:
    plt.subplot(4,4,j)
    sns.histplot(df[i], bins = 7, kde = True)
    plt.title(f'{i}')
    j=j+1
plt.show()
```

- **From the count plot, segment 2 has the highest number of occurrences, followed by segment 1 and then segment 3.**
- This suggests that wines belonging to segment 2 are the most common in this dataset.
- If this dataset is used for classification, the model might be slightly biased toward predicting segment 2 more often.
